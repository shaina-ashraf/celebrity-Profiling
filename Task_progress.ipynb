```python
import pandas
import os
```

# Task Verifier

## Fatima 

```python
fatima = pandas.read_excel('Work Division/Fatima.xlsx')
fatima.dropna(inplace=True)
fatima.usernames = fatima.usernames.str.replace('@','')
fatima.set_index('usernames',inplace=True)
```

```python
for path, directory, elements in os.walk('output_files/'):    
    for element in elements:                
        if '.csv' in element :            
            try:                
                fatima.loc[element.replace('.csv',''),'File Status'] = "File Found"
            except:
                print(element,'no such file found')            
```

```python
fatima.loc[fatima['File Status']!='File Found'].to_excel('fatima_remaining.xlsx')
```
