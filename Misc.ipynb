```python
data = pandas.read_excel('output_files/dataset/Sample data/sample_labels.xlsx')

data.Birthyear = data.Birthyear.apply(lambda x: x[1:5] if type(x)==str else 'NaN')
data['bins'] = pandas.cut(data['Birthyear'].str,bins=[1949,1980,2000],labels=['one','two','three'])
```
